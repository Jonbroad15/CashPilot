```python
import pandas as pd
```

```python
df = pd.read_csv('../master.csv')
df.head()
```

```python
print(f'Length of df: {len(df)}')
df = df.drop_duplicates(subset=['date', 'description', 'debit', 'credit'])
print(f'Length of df after dropping duplicates: {len(df)}')
```

```python
# Clean up account column
df = df.rename(columns={'Account': 'account'})
df['account'] = df['account'].apply(lambda x: str(x).lower().replace('.csv', ''))
```

```python
df['account'] = df['account'].apply(lambda x: x.replace('reserve', 'aeroplan'))
```

```python
set(df['account'])
```

```python
df.loc[df['date'] < '2024-01-14', ['account']] = 'cibc_dividend'
```

```python
df.to_csv('../master.csv', index=False)
```
